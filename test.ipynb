```python
import hashlib

def file_hash_from_path(path, chunk_size=8192):
    sha = hashlib.sha256()
    with open(path, "rb") as f:
        for chunk in iter(lambda: f.read(chunk_size), b""):
            sha.update(chunk)
    return sha.hexdigest()


file_hash_from_path("sms_screenshot/image_1.png")
```

```python
def convert_bytes_to_hash(file):
    sha = hashlib.sha256()

    # read file as bytes
    file_bytes = file.getvalue()

    sha.update(file_bytes)

    return sha.hexdigest()


```
